```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
hero_df = pd.read_csv(file_to_load)
hero_df.head()
```

## Player Count

```python
total_players = len(hero_df["SN"].value_counts())
total_players
```

```python
df_players = pd.DataFrame({"Total Players": [total_players]})
df_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
## Purchasing Analysis (Total)
unique_items = len(hero_df["Item ID"].value_counts())
unique_items

```

```python
#Total Rev
total_revenue = hero_df["Price"].sum()
total_revenue
```

```python
# Purchases
number_of_purchases = len(hero_df["Purchase ID"].value_counts())
number_of_purchases
```

```python
avg_purchases= hero_df["Price"].mean()
avg_purchases
```

```python
#DF Analysis Purchasing
df_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                           "Average Price": [avg_purchases],
                           "Number of Purchases": [number_of_purchases],
                           "Total Revenue": [total_revenue]})
df_analysis["Average Price"]=df_analysis["Average Price"].astype(float).map("${:,.2f}".format)
df_analysis["Total Revenue"]=df_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
df_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Demographics
#Establish bins for ages <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins =[0, 9, 14, 19, 24, 29, 34, 39, 999]
bins_for_ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
#Categorize the existing players using the age bins
hero_df["Categorized players"]= pd.cut(hero_df["Age"],bins, labels=bins_for_ages)
#Calculate the numbers and percentages by age group
age_grouped = hero_df.groupby("Categorized players")
count_age = age_grouped["SN"].nunique()
percentage_by_age = (count_age/total_players) * 100
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": count_age})
age_demographics.index.name = None
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Purchasing Analysis (Age)
purchase_age_count = age_grouped["Purchase ID"].count()
avg_purchase=age_grouped["Price"].mean()
total_purchase_age = age_grouped["Price"].sum()
avg_purchase_age = (total_purchase_age/count_age)*100
age_analysis_df = pd.DataFrame({"Purchase Count": purchase_age_count,
                               "Average Purchase Price": avg_purchase,
                                "Total Purchase Price": total_purchase_age,
                               "Avg Total Purchase per Person":avg_purchase_age})
age_analysis_df["Average Purchase Price"]= age_analysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_analysis_df["Total Purchase Price"]= age_analysis_df["Total Purchase Price"].astype(float).map("${:,.2f}".format)
age_analysis_df["Avg Total Purchase per Person"]= age_analysis_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
buyer = hero_df.groupby("SN")
count_buyers = buyer["Purchase ID"].count()
avg_buyers = buyer["Price"].mean()
purchase_total_buyers = buyer["Price"].sum()
top_spender_df = pd.DataFrame({"Purchase Count": count_buyers,
                              "Average Purchase Price": avg_buyers,
                              "Total Purchase Value": purchase_total_buyers})
filtered_top_spender_df = top_spender_df.sort_values(["Total Purchase Value"], ascending =False)
filtered_top_spender_df["Average Purchase Price"]= top_spender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
filtered_top_spender_df["Total Purchase Value"]= top_spender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_top_spender_df
```

```python
#most  Popular Items

items =hero_df[["Item ID", "Item Name", "Price"]]
items_grouped = hero_df.groupby(["Item ID", "Item Name"])
purchase_count = items_grouped["Price"].count()
purchase_price = items_grouped["Price"].mean()
price_per_item = purchase_price/purchase_count
popular_df = pd.DataFrame({"Purchase Count": purchase_count,
                          "Item Price": price_per_item,
                          "Total Purchase Value": purchase_price})
filtered_popular_df = popular_df.sort_values(["Purchase Count"],ascending = False)
filtered_popular_df["Item Price"]= popular_df["Item Price"].astype(float).map("${:,.2f}".format)
filtered_popular_df["Total Purchase Value"]= popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_popular_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#most  Popular Items

items =hero_df[["Item ID", "Item Name", "Price"]]
items_grouped = hero_df.groupby(["Item ID", "Item Name"])
purchase_count = items_grouped["Price"].count()
purchase_price = items_grouped["Price"].sum()
price_per_item = purchase_price/purchase_count
popular_df = pd.DataFrame({"Purchase Count": purchase_count,
                          "Item Price": price_per_item,
                          "Total Purchase Value": purchase_price})
filtered_popular_df= popular_df.sort_values(["Purchase Count"], ascending = False)
filtered_popular_df["Item Price"]= popular_df["Item Price"].astype(float).map("${:,.2f}".format)
filtered_popular_df["Total Purchase Value"]=popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_popular_df.head(10)
```
